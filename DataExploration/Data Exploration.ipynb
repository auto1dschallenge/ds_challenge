### Check distribution of price vs categorical features
Results could be used for feature engineering/selection

```python

cat_cols = ["make","body-style","drive-wheels","engine-type","fuel-system",
            "fuel-type","aspiration","num-of-doors","engine-location",'num-of-cylinders']
for i in cat_cols:
    ax = sns.boxplot(x=i, y="price", data=df)
    plt.title(i)
    plt.show()
```

#### Notes:
* high cardinality of make
* fwd and 4wd drive wheels have very similar distribution thats different from rwd
* mpfi fuel system has a high number of outliers with very high prices
* num-of-cylinders is basically a continous feature

### Check distribution of price vs categorical features
Results could be used for feature engineering/selection

```python
cont_cols = ["symboling","normalized-losses","wheel-base","length","width",
            "height","curb-weight","engine-size","bore","stroke","compression-ratio",
           "horsepower","peak-rpm","city-mpg","highway-mpg","price"]
for i in cont_cols:
    ax = sns.regplot(x=i, y="price", data=df,fit_reg=True)
    plt.title(i)
    plt.show()
```

#### Notes:
* symboling and normalized-losses have little clear influence on price. This makes sense given how both should be normalized to price/similar car type
* car hight with very little correlation to price as opposed to length and widt
* stroke and compression ratio with small correlation

All statements on correlation just based on visuals

### Systematically check correlations between cont features and price

```python
corr_df=df[cont_cols ] #New dataframe to calculate correlation between numeric features
cor= corr_df.corr(method='pearson')

```

```python
fig, ax =plt.subplots(figsize=(8, 6))
plt.title("Abs Correlation Plot")
sns.heatmap(np.abs(cor), mask=np.zeros_like(cor, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),square=True, ax=ax)
plt.show()
```

### Feature selection based on correlation
Arbitrarily define a cutoff value of 0.5 and only include features whose abs(correlation) with price is above this value to include in the model.

```python
rel_cont_cols=cor.price[cor.price<0.5].index.tolist()
print(rel_cont_cols)
```
